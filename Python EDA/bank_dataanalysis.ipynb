```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\vishn\Downloads\Box\End to End Projects\Banking\Banking.csv")
df.head()
```

```python
bins=[0,100000,300000,float("inf")]
labels=["Low","Med","High"]
df["Income Band"]=pd.cut(df["Estimated Income"],bins=bins,labels=labels,right=False)
df.head()
```

```python
df["Income Band"].value_counts()
```

```python
df["Income Band"].value_counts().plot(kind="bar")
```

```python
categorical_cols=df[["BRId","GenderId","IAId","Nationality","Occupation","Fee Structure","Loyalty Classification","Income Band","Properties Owned","Risk Weighting"]].columns

for col in categorical_cols:
    print(f"Value counts for '{col}' : ")
    display(df[col].value_counts())
    print("============================")
```

```python
#Univariate Analysis
for i,predictor in enumerate(df[["BRId","GenderId","IAId","Nationality","Occupation","Fee Structure","Loyalty Classification","Income Band","Properties Owned","Risk Weighting"]]):
    plt.figure(i)
    sns.countplot(data=df,x=predictor)    
```

```python
#Bivariate Analysis
for i,predictor in enumerate(df[["BRId","GenderId","IAId","Nationality","Occupation","Fee Structure","Loyalty Classification","Income Band","Properties Owned","Risk Weighting"]]):
    plt.figure(i)
    sns.countplot(data=df,x=predictor, hue="GenderId") 
```

```python
for i,predictor in enumerate(df[["BRId","GenderId","IAId","Nationality","Occupation","Fee Structure","Loyalty Classification","Income Band","Properties Owned","Risk Weighting"]]):
    plt.figure(i)
    sns.countplot(data=df,x=predictor, hue="Nationality")
```

```python
# Histplot of value counts for different Occupation
for col in categorical_cols:
    if col == "Occupation":
        continue
    plt.figure(figsize=(8,6))
    sns.histplot(df[col])
    plt.title('Histogram of Occupation Count')
    plt.xlabel(col)
    plt.ylabel("Count")
plt.show()
```

```python
numerical_cols=["Estimated Income","Superannuation Savings","Credit Card Balance","Bank Loans","Bank Deposits","Checking Accounts","Saving Accounts","Foreign Currency Account","Business Lending"]
plt.figure(figsize=(15,15))
for i ,col in enumerate(numerical_cols):
    plt.subplot(3,3,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
plt.show()    
```

```python
numerical_cols=["Estimated Income","Superannuation Savings","Credit Card Balance","Bank Loans","Bank Deposits","Checking Accounts","Saving Accounts","Foreign Currency Account","Business Lending"]
plt.figure(figsize=(8,8))
correlation_matrix=df[numerical_cols].corr()
sns.heatmap(correlation_matrix,annot=True,cmap="crest",fmt=".2f")
plt.title("correlation_matrix")
plt.show()
```
